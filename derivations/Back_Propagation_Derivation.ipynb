# Back Prop

## Main Rules for Back Prop

Thinking tips

- The derivative of something (A) to something else (B), put the result's dimension aside, is basially the **collection of individual element in A 's derivative to individual element in B**.

  - For example, if A is a scalar, B is a vector (b1, b2, b3), then the result is a collection of (dA/db1, dA/db2, dA/db3). (ignore the dimension)

- For the direvative of a single element (L) to another single element (w_ij), use this chain rule

Notations for the derivations

- x is the **input** to the whole network. Shape = (1, lx); e.g. (1, 2).
- w is the **weight** of the linear layer. Shape = (lw1, lw2)
- a is the **activation** from the hidden layers. Shape = (1, la)
- sigma is the **activation fn**, e.g. sigmoid.

### 5. The derivative of the weight to the linear layer is the **outer product** of the input (x, or a) and the backprop value to this point.

Note about **inner product** & **outer product**

- The inner product (also known as the dot product) and outer product are two fundamental operations in linear algebra that involve multiplying **two vectors**. The inner product results in a scalar (a single number), while the outer product results in a matrix.
